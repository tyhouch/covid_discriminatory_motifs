```python
import pandas as pd
import sys, re
import numpy as np
import seaborn as sns
```

```python
#reading in our motifs and saving as an array motifs['k_mer'] = 'ACTG......'
f=open('k_mers/kmers.fasta','r')
lines=f.readlines()

hre=re.compile('>(\S+)')
lre=re.compile('^(\S+)$')

#creates an array gene['ID'] = 'ACTG......'
motifs={}

for line in lines:
        outh = hre.search(line)
        if outh:
                id=outh.group(1)
        else:
                outl=lre.search(line)
                if(id in motifs.keys()):
                        motifs[id] += outl.group(1)
                else:
                        motifs[id] = outl.group(1)
```

```python
#creates an array of motifs
motif_array = []
for m in motifs:
    motif_array.append(motifs[m]) 
motif_array = np.array(motif_array)
```
